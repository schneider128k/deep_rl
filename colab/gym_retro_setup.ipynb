** You need to obtain the rom files for the Atari games yourself because they are not provided automatically when you install gym-retro. **
1. Download Rom.rar file from [Atarimania](http://www.atarimania.com/rom_collection_archive_atari_2600_roms.html).
2. Extract Roms.rar into Rom folder. The Roms folder contains now two files: HC ROMS.rar and ROMS.rar
3. Extract ROMS.rar into ROMS folder. The ROMS folder contains now the rom files for many Atari games.
3. Upload the ROMS folder to your Google drive.
