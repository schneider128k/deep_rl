```python
import numpy as np
from matplotlib import pyplot as plt
import os
from IPython import display
```

```python
DIR_NAME = './data/rollout/'
file_index = 10
file = os.listdir(DIR_NAME)[file_index]

rollout = np.load(DIR_NAME + file)
obs_data = rollout['obs']
action_data = rollout['action']
reward_data = rollout['reward']

print('observations', obs_data.shape)
print('actions', action_data.shape)
print('rewards', reward_data.shape)
```

```python
fig, axes = plt.subplots(nrows=5, ncols=4, figsize=(64, 64), sharex=True, sharey=True)
plt.subplots_adjust(hspace = 0.4)
idx = 0
for x in range(5):
    for y in range(4):
        
        obs = obs_data[idx]
        axes[x, y].imshow(obs)
        
        axes[x, y].text(0.5, -0.1, 
                        'frame ' + str(idx), 
                        size=40, 
                        ha='center', 
                        transform=axes[x,y].transAxes)
        
        axes[x, y].text(0.5, -0.2, 
                        'a = ' + str(np.round(action_data[idx], 2)), 
                        size=40, 
                        ha='center', 
                        transform=axes[x,y].transAxes)
        
        axes[x, y].text(0.5, -0.3, 
                        'r = ' + str(np.round(reward_data[idx + 1], 2)), 
                        size=40, 
                        ha='center', 
                        transform=axes[x,y].transAxes)
        
        idx += 15

plt.show()
```
